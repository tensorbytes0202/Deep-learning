```python
# =============================
# 🔧 INSTALL DEPENDENCIES
# =============================
!apt-get install -y tesseract-ocr
!pip install pytesseract opencv-python matplotlib numpy

# =============================
# 📥 IMPORT LIBRARIES
# =============================
import cv2
import numpy as np
import pytesseract
import matplotlib.pyplot as plt
import re

# =============================
# 📂 LOAD IMAGE
# =============================
image_path = "/content/62.png"  # upload your image

img = cv2.imread(image_path)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')
```

```python
# Threshold
_, thresh = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY_INV)

# Morphology (clean noise)
kernel = np.ones((5,5), np.uint8)
closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

# Find contours (rooms)
contours, _ = cv2.findContours(closing, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

mask = np.zeros_like(gray)

room_contours = []

for cnt in contours:
    area = cv2.contourArea(cnt)

    if area > 2000:  # filter small noise
        room_contours.append(cnt)
        cv2.drawContours(mask, [cnt], -1, 255, -1)

plt.imshow(mask, cmap='gray')
plt.title("Segmentation Mask")
plt.axis('off')

print("Estimated rooms:", len(room_contours))
```

```python
# Edge detection
edges = cv2.Canny(gray, 50, 150)

# Hough circle (doors often arc-like)
circles = cv2.HoughCircles(
    gray,
    cv2.HOUGH_GRADIENT,
    dp=1.2,
    minDist=50,
    param1=50,
    param2=30,
    minRadius=10,
    maxRadius=60
)

door_count = 0

if circles is not None:
    door_count = len(circles[0])

print("Estimated Doors:", door_count)
```

```python
# Windows = small rectangular gaps in walls

contours, _ = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

window_count = 0

for cnt in contours:
    x, y, w, h = cv2.boundingRect(cnt)

    if 20 < w < 100 and 5 < h < 40:
        window_count += 1

print("Estimated Windows:", window_count)
```

```python
# OCR
custom_config = r'--oem 3 --psm 6'
text = pytesseract.image_to_string(gray, config=custom_config)

print("Raw OCR Text:\n", text)
```

```python
# Regex for area (e.g. 22.96 m²)
pattern = r'([A-Z]+)\s*([\d\.]+)\s*m'

matches = re.findall(pattern, text)

rooms = {}

for name, area in matches:
    rooms[name] = float(area)

print("\nDetected Rooms & Sizes:")
for k, v in rooms.items():
    print(f"{k} : {v} m²")
```

```python
print("\n========== FINAL OUTPUT ==========")
print("Rooms detected:", len(rooms))
print("Doors:", door_count)
print("Windows:", window_count)

print("\nRoom Details:")
for room, size in rooms.items():
    print(f"{room}: {size} m²")
```

```python
# =============================
# BETTER ROOM SEGMENTATION
# =============================

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread(image_path)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Step 1: Threshold
_, thresh = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY_INV)

# Step 2: Remove noise
kernel = np.ones((3,3), np.uint8)
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=2)

# Step 3: Sure background
sure_bg = cv2.dilate(opening, kernel, iterations=3)

# Step 4: Distance transform
dist = cv2.distanceTransform(opening, cv2.DIST_L2, 5)

# Step 5: Sure foreground (rooms)
_, sure_fg = cv2.threshold(dist, 0.4*dist.max(), 255, 0)

sure_fg = np.uint8(sure_fg)

# Step 6: Unknown region
unknown = cv2.subtract(sure_bg, sure_fg)

# Step 7: Marker labeling
_, markers = cv2.connectedComponents(sure_fg)

markers = markers + 1
markers[unknown == 255] = 0

# Step 8: Watershed
markers = cv2.watershed(img, markers)

# Step 9: Visualization
mask = np.zeros_like(gray)

for label in np.unique(markers):
    if label <= 1:
        continue
    mask[markers == label] = 255

plt.imshow(mask, cmap='gray')
plt.title("Improved Room Segmentation")
plt.axis('off')

print("Rooms detected:", len(np.unique(markers)) - 2)
```

```python
gray = cv2.GaussianBlur(gray, (5,5), 0)

custom_config = r'--oem 3 --psm 11'
text = pytesseract.image_to_string(gray, config=custom_config)

print(text)
```

```python
# Strong wall detection
edges = cv2.Canny(gray, 100, 200)

kernel = np.ones((7,7), np.uint8)
walls = cv2.dilate(edges, kernel, iterations=3)

plt.imshow(walls, cmap='gray')
plt.title("Walls (Strong)")
plt.axis('off')
```

```python
# Remove small noisy components (text)
contours, _ = cv2.findContours(walls, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

clean = np.zeros_like(walls)

for cnt in contours:
    area = cv2.contourArea(cnt)

    if area > 500:  # ignore text
        cv2.drawContours(clean, [cnt], -1, 255, -1)

plt.imshow(clean, cmap='gray')
plt.title("Clean Walls")
plt.axis('off')
```

```python
# Invert
inv = cv2.bitwise_not(clean)

# Flood fill background
h, w = inv.shape
mask_ff = np.zeros((h+2, w+2), np.uint8)

flood = inv.copy()
cv2.floodFill(flood, mask_ff, (0,0), 0)

# Remaining white = rooms
rooms = flood

plt.imshow(rooms, cmap='gray')
plt.title("Rooms (Isolated)")
plt.axis('off')
```

```python
num_labels, labels = cv2.connectedComponents(rooms)

final_mask = np.zeros_like(gray)
room_count = 0

for label in range(1, num_labels):
    area = np.sum(labels == label)

    if 8000 < area < 200000:  # strict filtering
        final_mask[labels == label] = 255
        room_count += 1

plt.imshow(final_mask, cmap='gray')
plt.title("Final Rooms")
plt.axis('off')

print("Rooms:", room_count)
```
