```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
```

```python
(x_train,y_train),(x_test,y_test) = tf.keras.datasets.mnist.load_data()
```

```python
x_train , x_test= x_train/255.0,x_test/255.0
#scale 0-255 -> 0-1(Fast convergence)
```

```python
#feedforward fully connected neural network  used for multiclass classification
mlp = models.Sequential([
    #layers ko line by line stack karta h
    layers.Flatten(input_shape=(28,28)),
    #2D image -> 1D vector(28*28 ->784)
    layers.Dense(128,activation='relu'),
    #for conceting the all the neuron of previous and next layer
    layers.Dense(64,activation='relu'),
    layers.Dense(10,activation='softmax')
    #for multi-class prob output

])


#vanishing gradient -> "Gradients become very small during backpropagation, causing early layers to stop learning."
# max(0,x)
```

```python
mlp.compile(optimizer='adam',
            #improve gradient descent,adaptive learning for every weights
            #it combines momentum(past gradients) and RMS(squared gradients) and then updatees the weight
            loss='sparse_categorical_crossentropy',
           #multi-class classification loss function,in the labels integer
            metrics=['accuracy'])
```

```python
mlp.fit(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

```python
# -------------------------CNN MODEL--------------------------------------------------
x_train_cnn = x_train[..., tf.newaxis]
x_test_cnn = x_test[..., tf.newaxis]

# channel req for cnn training (batch_size, height, width, channels)
#(60000, 28, 28, 1) like this 1 shows grayscale image
```

```python

#for extracting spatial features from the image(edges,shapes)
cnn = models.Sequential([
    layers.Conv2D(32,(3,3),activation='relu',input_shape=(28,28,1)),
    #32 filters → 32 different features learn karega
    #(3×3 kernel) → small window scan karega image pe
    #Image pe sliding window chala ke features detect karta hai:
    # relu for Non-linearity add
    layers.MaxPooling2D(2,2),
    #MaxPooling reduces spatial dimensions while retaining important features
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D(2,2),
    layers.Flatten(),
    layers.Dense(128,activation='relu'),
    layers.Dense(10,activation='softmax')
])
```

```python
cnn.compile(optimizer='adam',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
            )
```

```python
cnn.fit(x_train_cnn, y_train, epochs=5, validation_data=(x_test_cnn,y_test))
```
