Phase 2 Resizeing and scaling

```python
# Load your image
image = cv2.imread("IMG-20250824-WA0003.jpg")  # change to your path

if image is None:
    print("Image not found!")
else:
    print(" Image loaded successfully.")

    # Resize image
    resized = cv2.resize(image, (5000, 5000))

    # Convert BGR  RGB (Matplotlib displays images in RGB order)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    resized_rgb = cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)

    # Show both images side by side
    plt.figure(figsize=(8, 4))

    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(image_rgb)
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Resized Image")
    plt.imshow(resized_rgb)
    plt.axis('off')

    plt.show()

    # Save the resized image
    cv2.imwrite("resized.png", resized)
    print(" Resized image saved as 'resized.png'")
```

```python
if image is not None:
    # Crop the image
    cropped = image[200:800, 20:150]
    
    # Convert BGR to RGB for correct colors in matplotlib
    cropped_rgb = cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB)
    
    # Display using matplotlib
    plt.imshow(cropped_rgb)
    plt.axis('off')  # Hide axes
    plt.show()
    
    # Save the cropped image
    cv2.imwrite("cropped.png", cropped)
```

```python

if image is None:
    print("Could not load image")
else:
    # Get image dimensions
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    
    # Rotation matrix (90 degrees clockwise)
    M = cv2.getRotationMatrix2D(center, 90, 1.0)
    rotated = cv2.warpAffine(image, M, (w, h))
    
    # Convert BGR to RGB for correct colors
    rotated_rgb = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)
    
    # Display using matplotlib
    plt.imshow(rotated_rgb)
    plt.axis('off')
    plt.show()
    
    # Optionally save the rotated image
    cv2.imwrite("rotated.png", rotated)
```

```python

if image is None:
    print("Could not load image")
else:
    # Flip horizontally
    flipped_horizontal = cv2.flip(image, 1)
    # Flip vertically
    flipped_vertical = cv2.flip(image, 0)
    # Flip both axes
    flipped_both = cv2.flip(image, -1)
    
    # Helper function to display images with matplotlib
    def show_image(img, title="Image"):
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img_rgb)
        plt.axis('off')
        plt.title(title)
        plt.show()
    
    # Display all flipped versions
    show_image(flipped_horizontal, "Flipped Horizontal")
    show_image(flipped_vertical, "Flipped Vertical")
    show_image(flipped_both, "Flipped Both")
    
    # Optionally save the flipped images
    cv2.imwrite("flipped_horizontal.png", flipped_horizontal)
    cv2.imwrite("flipped_vertical.png", flipped_vertical)
    cv2.imwrite("flipped_both.png", flipped_both)
```

```python

# Assuming 'image' is already read, e.g., image = cv2.imread("your_image.png")

if image is None:
    print("Could not load image")
else:
    print("Image loaded Successfully!")
    
    # Define points, color, and thickness
    pt1 = (440, 800)
    pt2 = (200, 100)
    color = (0, 0, 255)  # BGR (red)
    thickness = 50
    
    # Draw the line
    cv2.line(image, pt1, pt2, color, thickness)
    
    # Convert BGR  RGB for correct colors in matplotlib
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Display using matplotlib
    plt.imshow(image_rgb)
    plt.axis('off')
    plt.show()
    
    # Optionally save the image
    cv2.imwrite("line_drawn.png", image)
```
