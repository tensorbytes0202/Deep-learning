Generic Protocol Parser

```python
# ==============================================================
# Generic Protocol Parser
#
# Supports:
#   ASVspoof2019
#   ASVspoof2021
#   ASVspoof2025 (editable)
# ==============================================================

import pandas as pd
from pathlib import Path


class ProtocolParser:

    """
    Generic Protocol Parser

    Output Columns

    audio_id
    speaker
    attack
    label
    path
    """

    def __init__(
        self,
        protocol_path,
        audio_dir,
        extension=".flac"
    ):

        self.protocol_path = Path(protocol_path)
        self.audio_dir = Path(audio_dir)
        self.extension = extension

        self.data = self._read_protocol()

    # ---------------------------------------------------------

    def _read_protocol(self):

        samples = []

        with open(
            self.protocol_path,
            "r",
            encoding="utf-8"
        ) as f:

            for line in f:

                line = line.strip()

                if len(line) == 0:
                    continue

                tokens = line.split()

                # -----------------------------------------
                # ASVspoof2019
                #
                # speaker
                # audio_id
                # -
                # attack
                # label
                # -----------------------------------------

                if len(tokens) == 5:

                    speaker = tokens[0]

                    audio_id = tokens[1]

                    attack = tokens[3]

                    label = tokens[4]

                # -----------------------------------------
                # ASVspoof2025
                #
                # Change Here
                # -----------------------------------------

                else:

                    speaker = "unknown"

                    audio_id = tokens[0]

                    attack = "unknown"

                    label = tokens[-1]

                audio_path = (
                    self.audio_dir /
                    f"{audio_id}{self.extension}"
                )

                samples.append(

                    {

                        "speaker": speaker,

                        "audio_id": audio_id,

                        "attack": attack,

                        "label": label,

                        "path": audio_path

                    }

                )

        return pd.DataFrame(samples)

    # ---------------------------------------------------------

    def dataframe(self):

        return self.data

    # ---------------------------------------------------------

    def statistics(self):

        print("=" * 60)

        print("Total Samples :", len(self.data))

        print()

        print("Labels")

        print(self.data["label"].value_counts())

        print()

        print("Attacks")

        print(self.data["attack"].value_counts())

        print("=" * 60)

    # ---------------------------------------------------------

    def validate_paths(self):

        print("=" * 60)

        total = len(self.data)

        missing = 0

        for p in self.data["path"]:

            if not Path(p).exists():

                missing += 1

        print("Total :", total)

        print("Missing :", missing)

        print("Found :", total - missing)

        print("=" * 60)
```

Audio Processing Engine

```python
# ============================================================
# Audio Processing Engine
# ============================================================

class AudioProcessor:

    def __init__(
        self,
        sample_rate=16000,
        max_length=64000
    ):

        self.sample_rate = sample_rate
        self.max_length = max_length

    # --------------------------------------------------------
    # Load Audio
    # --------------------------------------------------------

    def load_audio(self, path):

        audio, sr = sf.read(path)

        return audio, sr

    # --------------------------------------------------------
    # Stereo -> Mono
    # --------------------------------------------------------

    def to_mono(self, audio):

        if len(audio.shape) == 2:

            audio = np.mean(audio, axis=1)

        return audio

    # --------------------------------------------------------
    # Resample
    # --------------------------------------------------------

    def resample(self, audio, sr):

        if sr != self.sample_rate:

            audio = librosa.resample(

                audio,

                orig_sr=sr,

                target_sr=self.sample_rate

            )

        return audio

    # --------------------------------------------------------
    # Remove Silence
    # --------------------------------------------------------

    def trim_silence(self, audio):

        audio, _ = librosa.effects.trim(

            audio,

            top_db=25

        )

        return audio

    # --------------------------------------------------------
    # RMS Normalization
    # --------------------------------------------------------

    def rms_normalize(self, audio):

        rms = np.sqrt(np.mean(audio**2))

        if rms > 0:

            audio = audio / rms

        return audio

    # --------------------------------------------------------
    # Peak Normalization
    # --------------------------------------------------------

    def peak_normalize(self, audio):

        peak = np.max(np.abs(audio))

        if peak > 0:

            audio = audio / peak

        return audio

    # --------------------------------------------------------
    # Random Crop
    # --------------------------------------------------------

    def random_crop(self, audio):

        if len(audio) <= self.max_length:

            return audio

        start = random.randint(

            0,

            len(audio)-self.max_length

        )

        return audio[
            start:
            start+self.max_length
        ]

    # --------------------------------------------------------
    # Center Crop
    # --------------------------------------------------------

    def center_crop(self, audio):

        if len(audio)<=self.max_length:

            return audio

        start=(len(audio)-self.max_length)//2

        return audio[
            start:
            start+self.max_length
        ]

    # --------------------------------------------------------
    # Padding
    # --------------------------------------------------------

    def pad(self,audio):

        if len(audio)>=self.max_length:

            return audio

        pad=self.max_length-len(audio)

        audio=np.pad(

            audio,

            (0,pad),

            mode="constant"

        )

        return audio

    # --------------------------------------------------------
    # Tensor
    # --------------------------------------------------------

    def to_tensor(self,audio):

        return torch.tensor(

            audio,

            dtype=torch.float32

        )
```

Dataset Class

```python
# ============================================================
# ASVspoof Dataset
# ============================================================

class ASVSpoofDataset(Dataset):

    def __init__(
        self,
        dataframe,
        processor,
        train=True
    ):

        self.df = dataframe.reset_index(drop=True)

        self.processor = processor

        self.train = train

    def __len__(self):

        return len(self.df)

    def __getitem__(self,index):

        sample=self.df.iloc[index]

        audio_path=sample["path"]

        label=LABEL2ID[sample["label"]]

        attack=sample["attack"]

        speaker=sample["speaker"]

        audio,sr=self.processor.load_audio(audio_path)

        audio=self.processor.to_mono(audio)

        audio=self.processor.resample(audio,sr)

        audio=self.processor.trim_silence(audio)

        audio=self.processor.rms_normalize(audio)

        audio=self.processor.peak_normalize(audio)

        if self.train:

            audio=self.processor.random_crop(audio)

        else:

            audio=self.processor.center_crop(audio)

        audio=self.processor.pad(audio)

        audio=self.processor.to_tensor(audio)

        return {

            "audio":audio,

            "label":torch.tensor(label),

            "speaker":speaker,

            "attack":attack,

            "audio_id":sample["audio_id"]

        }
```

```python
# ============================================================
# Custom Collate
# ============================================================

def collate_fn(batch):

    audio=[]

    label=[]

    speaker=[]

    attack=[]

    audio_id=[]

    for item in batch:

        audio.append(item["audio"])

        label.append(item["label"])

        speaker.append(item["speaker"])

        attack.append(item["attack"])

        audio_id.append(item["audio_id"])

    audio=torch.stack(audio)

    label=torch.stack(label)

    return {

        "audio":audio,

        "label":label,

        "speaker":speaker,

        "attack":attack,

        "audio_id":audio_id

    }
```

```python
# ============================================================
# Weighted Layer Fusion
# ============================================================

class WeightedLayerFusion(nn.Module):

    def __init__(self,n_layers=25):

        super().__init__()

        self.weights = nn.Parameter(

            torch.ones(n_layers)

        )

    def forward(self,hidden_states):

        w = F.softmax(

            self.weights,

            dim=0

        )

        output = 0

        for i in range(len(hidden_states)):

            output += w[i]*hidden_states[i]

        return output
```

FAISS Memory Bank

```python
# ============================================================
# FAISS Memory Bank
# ============================================================

import faiss
import numpy as np
import torch


class MemoryBank:

    def __init__(

        self,

        embedding_dim=256,

        use_gpu=False

    ):

        self.embedding_dim = embedding_dim

        self.use_gpu = use_gpu

        self.metadata = []

        self.index = faiss.IndexFlatIP(
            embedding_dim
        )

        if self.use_gpu:

            res = faiss.StandardGpuResources()

            self.index = faiss.index_cpu_to_gpu(

                res,

                0,

                self.index

            )
```

```python
# ------------------------------------------------------------
# Number of embeddings
# ------------------------------------------------------------

    def __len__(self):

        return self.index.ntotal
```

```python
# ------------------------------------------------------------
# Add Embeddings
# ------------------------------------------------------------

    def add(

        self,

        embeddings,

        metadata=None

    ):

        if torch.is_tensor(embeddings):

            embeddings = embeddings.detach().cpu().numpy()

        embeddings = embeddings.astype(np.float32)

        faiss.normalize_L2(embeddings)

        self.index.add(embeddings)

        if metadata is not None:

            self.metadata.extend(metadata)
```

```python
# ------------------------------------------------------------
# Search
# ------------------------------------------------------------

    def search(

        self,

        query,

        top_k=5

    ):

        if torch.is_tensor(query):

            query = query.detach().cpu().numpy()

        query = query.astype(np.float32)

        faiss.normalize_L2(query)

        scores, indices = self.index.search(

            query,

            top_k

        )

        return scores, indices
```

```python
# ------------------------------------------------------------
# Get Metadata
# ------------------------------------------------------------

    def get_metadata(

        self,

        indices

    ):

        result=[]

        for row in indices:

            temp=[]

            for idx in row:

                if idx==-1:

                    temp.append(None)

                else:

                    temp.append(

                        self.metadata[idx]

                    )

            result.append(temp)

        return result
```

```python
# ------------------------------------------------------------
# Save Memory
# ------------------------------------------------------------

    def save(

        self,

        file_name="memory.index"

    ):

        if self.use_gpu:

            cpu_index = faiss.index_gpu_to_cpu(

                self.index

            )

            faiss.write_index(

                cpu_index,

                file_name

            )

        else:

            faiss.write_index(

                self.index,

                file_name

            )

        print("Memory Saved")
```

```python
# ------------------------------------------------------------
# Load Memory
# ------------------------------------------------------------

    def load(

        self,

        file_name="memory.index"

    ):

        self.index = faiss.read_index(

            file_name

        )

        print("Memory Loaded")
```

Embedding Store + Retriever

```python
# ============================================================
# Embedding Store
# ============================================================

class EmbeddingStore:

    def __init__(self):

        self.embeddings = []

        self.labels = []

        self.audio_ids = []

        self.attacks = []

        self.speakers = []

    def add(

        self,

        embeddings,

        labels,

        audio_ids,

        attacks,

        speakers

    ):

        if torch.is_tensor(embeddings):

            embeddings = embeddings.detach().cpu()

        self.embeddings.append(embeddings)

        self.labels.extend(labels)

        self.audio_ids.extend(audio_ids)

        self.attacks.extend(attacks)

        self.speakers.extend(speakers)

    def build(self):

        self.embeddings = torch.cat(

            self.embeddings,

            dim=0

        )

        return self.embeddings

    def __len__(self):

        return len(self.embeddings)
```

```python
class Retriever:

    def __init__(

        self,

        memory,

        faiss_index

    ):

        self.memory = memory

        self.index = faiss_index
```

```python
    def retrieve(

        self,

        query,

        top_k=5

    ):

        score,idx = self.index.search(

            query,

            top_k

        )

        retrieved = []

        for row in idx:

            retrieved.append(

                self.memory[row]

            )

        retrieved = torch.stack(

            retrieved

        )

        return retrieved
```

```python
class CrossAttention(nn.Module):

    def __init__(

        self,

        dim=256,

        heads=4

    ):

        super().__init__()

        self.attn = nn.MultiheadAttention(

            embed_dim=dim,

            num_heads=heads,

            batch_first=True

        )

    def forward(

        self,

        query,

        memory

    ):

        output,weights = self.attn(

            query,

            memory,

            memory

        )

        return output
```

```python
# ============================================================
# Prediction Function
# ============================================================

@torch.no_grad()

def predict(loader):

    classifier.eval()

    projection.eval()

    fusion.eval()

    cross.eval()

    ffn.eval()

    y_true=[]

    y_pred=[]

    y_score=[]

    for batch in tqdm(loader):

        audio=batch["audio"].to(DEVICE)

        label=batch["label"].to(DEVICE)

        inputs=feature_extractor(

            audio.cpu().numpy(),

            sampling_rate=CFG.SAMPLE_RATE,

            return_tensors="pt",

            padding=True

        )

        inputs={

            k:v.to(DEVICE)

            for k,v in inputs.items()

        }

        outputs=wavlm(

            **inputs,

            output_hidden_states=True

        )

        emb=fusion(outputs.hidden_states)

        emb=projection(emb)

        emb=emb.mean(dim=1)

        emb=F.normalize(emb,dim=1)

        logits=classifier(emb)

        prob=torch.softmax(logits,dim=1)

        pred=prob.argmax(1)

        y_true.extend(label.cpu().numpy())

        y_pred.extend(pred.cpu().numpy())

        y_score.extend(prob[:,1].cpu().numpy())

    return y_true,y_pred,y_score
```
