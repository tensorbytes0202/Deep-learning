Thresholding is a process of converting a grayscale image into a binary image (black & white) based on a threshold value.

üëâ In simple words:
If pixel intensity > threshold ‚Üí make it white (255)
Else ‚Üí make it black (0)

```python
import cv2
img = cv2.imread("Screenshot 2025-10-21 204614.png",cv2.IMREAD_GRAYSCALE)
ret,thresh_img= cv2.threshold(img,90,255,cv2.THRESH_BINARY)
cv2.imshow("Orginal Image",img)
cv2.imshow("Edges",thresh_img)
cv2.waitKey(0)
cv2.destroyAllWindows()

"""
90-black
130- 255 white
180 - 255 white
50 - 0 black 

100,120,150
"""
```

These are logical operations performed pixel-by-pixel on two (or sometimes one) images ‚Äî just like logical AND, OR, XOR, NOT in mathematics.
They are extremely useful in masking, combining images, isolating objects, and ROI (region of interest) operations.

üß† What Are Bitwise Operations?

Each pixel in an image is represented by 8 bits (0‚Äì255 intensity).
OpenCV lets you perform bitwise logic between pixels of two images (or between an image and a mask).

‚öôÔ∏è 1. Available Functions
Function	Description
cv2.bitwise_and()	Performs AND operation on each pixel
cv2.bitwise_or()	Performs OR operation on each pixel
cv2.bitwise_xor()	Performs XOR operation on each pixel
cv2.bitwise_not()	Inverts all bits (NOT operation)

```python
"""
img1 and img2 height and width should be same
use only black and white

"""
import cv2
import numpy as np
img1 = np.zeros((300,300),dtype = 'uint8')
img2 = np.zeros((300,300),dtype = 'uint8')
cv2.circle(img1,(150,150),100,255,-1)
cv2.rectangle(img2,(100,100),(250,250),255,-1)
bitwise_and = cv2.bitwise_and(img1,img2)
bitwise_or = cv2.bitwise_or(img1,img2)
bitwise_not = cv2.bitwise_not(img1)
cv2.imshow("AND",bitwise_and)
cv2.imshow("OR",bitwise_or)
cv2.imshow("NOT",bitwise_not)
cv2.waitKey(0)
cv2.destroyAllWindows()

```
