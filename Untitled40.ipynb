## 2. Import Libraries

I am importing all the libraries required for building the model.

```python
# ==========================================================
# STEP 2 : IMPORT LIBRARIES
# ==========================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import os
import zipfile

from collections import Counter

from google.colab import drive

from tensorflow.keras.preprocessing.image import ImageDataGenerator

from skimage.feature import hog
from skimage.feature import local_binary_pattern

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import learning_curve

from sklearn.preprocessing import StandardScaler

from sklearn.decomposition import PCA

from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

from xgboost import XGBClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

import warnings
warnings.filterwarnings('ignore')
```

## 4. Extract Dataset

```python
# ==========================================================
# STEP 4 : EXTRACT DATASET
# ==========================================================

zip_path = "/content/drive/MyDrive/hymenoptera_data.zip"
extract_path = "/content/hymenoptera_data"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Dataset Extracted Successfully!")
```

## 5. Load Dataset

```python
# ==========================================================
# STEP 5 : LOAD DATASET
# ==========================================================

dataset_path = "/content/hymenoptera_data/hymenoptera_data/train"

images = []
labels = []

classes = os.listdir(dataset_path)

print("Classes Found:", classes)

for label, class_name in enumerate(classes):
    class_path = os.path.join(dataset_path, class_name)

    if not os.path.isdir(class_path):
        continue

    for file in os.listdir(class_path):
        img_path = os.path.join(class_path, file)

        img = cv2.imread(img_path)

        if img is not None:
            images.append(img)
            labels.append(label)

images = np.array(images, dtype=object)
labels = np.array(labels)

print("Total Images:", len(images))
print("Classes:", classes)
```

## 6. Check Data Imbalance

```python
# ==========================================================
# STEP 6 : CHECK DATA IMBALANCE
# ==========================================================

counter = Counter(labels)

print("\nClass Distribution:\n")

for i, count in counter.items():
    print(f"{classes[i]} : {count}")

plt.figure(figsize=(6,4))
sns.countplot(x=labels)
plt.xticks(
    ticks=range(len(classes)),
    labels=classes
)
plt.title("Class Distribution")
plt.xlabel("Classes")
plt.ylabel("Number of Images")
plt.show()
```

## 7. Image Preprocessing

```python
# ==========================================================
# STEP 7 : IMAGE PREPROCESSING
# ==========================================================

processed_images = []
IMG_SIZE = 64

for img in images:
    # Resize
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    # Noise Reduction
    img = cv2.GaussianBlur(img, (3,3), 0)
    # Normalize
    img = img.astype("float32") / 255.0
    processed_images.append(img)

processed_images = np.array(processed_images)
print("Processed Shape:", processed_images.shape)
```

## 8. Image Augmentation

```python
# ==========================================================
# STEP 8 : IMAGE AUGMENTATION
# ==========================================================

datagen = ImageDataGenerator(
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.1,
    height_shift_range=0.1
)

datagen.fit(processed_images)
print("Augmentation Applied")
```

## 9. Feature Extraction

```python
# ==========================================================
# STEP 9 : FEATURE EXTRACTION
# ==========================================================

features = []

for img in processed_images:
    temp = (img * 255).astype("uint8")

    # ======================================================
    # COLOR FEATURE
    # ======================================================

    color_feature = cv2.calcHist(
        [temp],
        [0,1,2],
        None,
        [8,8,8],
        [0,256,0,256,0,256]
    ).flatten()

    # ======================================================
    # GRAYSCALE
    # ======================================================

    gray = cv2.cvtColor(temp, cv2.COLOR_BGR2GRAY)

    # ======================================================
    # TEXTURE FEATURE -> LBP
    # ======================================================

    lbp = local_binary_pattern(
        gray,
        24,
        3,
        method='uniform'
    )

    hist, _ = np.histogram(
        lbp.ravel(),
        bins=np.arange(0,27),
        range=(0,26)
    )

    hist = hist.astype("float")
    hist /= (hist.sum() + 1e-6)
    texture_feature = hist

    # ======================================================
    # SHAPE FEATURE -> HOG
    # ======================================================

    shape_feature = hog(
        gray,
        orientations=9,
        pixels_per_cell=(8,8),
        cells_per_block=(2,2),
        transform_sqrt=True,
        feature_vector=True
    )

    # ======================================================
    # COMBINE FEATURES
    # ======================================================

    final_feature = np.hstack([
        color_feature,
        texture_feature,
        shape_feature
    ])

    features.append(final_feature)

features = np.array(features)
print("Feature Shape:", features.shape)
```

## 10. Train Test Split

```python
# ==========================================================
# STEP 10 : TRAIN TEST SPLIT
# ==========================================================

X_train, X_test, y_train, y_test = train_test_split(
    features,
    labels,
    test_size=0.2,
    random_state=42,
    stratify=labels
)
```

## 11. Feature Scaling

```python
# ==========================================================
# STEP 11 : FEATURE SCALING
# ==========================================================

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## 12. PCA Dimension Reduction

```python
# ==========================================================
# STEP 12 : PCA DIMENSION REDUCTION
# ==========================================================

pca = PCA(n_components=0.95)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)

print("PCA Applied")
print("Reduced Shape:", X_train.shape)
```

## 13. SVM Model

```python
# ==========================================================
# STEP 13 : SVM MODEL
# ==========================================================

svm_params = {
    'C': [10, 100],
    'gamma': [0.001, 0.0001],
    'kernel': ['rbf']
}

svm_grid = GridSearchCV(
    SVC(class_weight='balanced'),
    svm_params,
    cv=3,
    scoring='accuracy'
)

svm_grid.fit(X_train, y_train)

best_svm = svm_grid.best_estimator_

svm_predictions = best_svm.predict(X_test)

svm_accuracy = accuracy_score(y_test, svm_predictions)

print("\nBest SVM Parameters:", svm_grid.best_params_)
print("SVM Accuracy:", svm_accuracy)
```

## 14. KNN Model

```python
# ==========================================================
# STEP 14 : KNN MODEL
# ==========================================================

knn_params = {
    'n_neighbors': [3,5,7],
    'weights': ['uniform', 'distance']
}

knn_grid = GridSearchCV(
    KNeighborsClassifier(),
    knn_params,
    cv=3,
    scoring='accuracy'
)

knn_grid.fit(X_train, y_train)

best_knn = knn_grid.best_estimator_

knn_predictions = best_knn.predict(X_test)

knn_accuracy = accuracy_score(y_test, knn_predictions)

print("\nBest KNN Parameters:", knn_grid.best_params_)
print("KNN Accuracy:", knn_accuracy)
```

## 15. Random Forest Model

```python
# ==========================================================
# STEP 15 : RANDOM FOREST
# ==========================================================

rf_params = {
    'n_estimators': [100, 200],
    'max_depth': [20, 30],
    'min_samples_split': [2,5]
}

rf_grid = GridSearchCV(
    RandomForestClassifier(),
    rf_params,
    cv=3,
    scoring='accuracy'
)

rf_grid.fit(X_train, y_train)

best_rf = rf_grid.best_estimator_

rf_predictions = best_rf.predict(X_test)

rf_accuracy = accuracy_score(y_test, rf_predictions)

print("\nBest RF Parameters:", rf_grid.best_params_)
print("RF Accuracy:", rf_accuracy)
```

## 16. XGBoost Model

```python
# ==========================================================
# STEP 16 : XGBOOST MODEL
# ==========================================================

xgb_model = XGBClassifier(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective='binary:logistic',
    eval_metric='logloss',
    random_state=42
)

xgb_model.fit(X_train, y_train)

xgb_predictions = xgb_model.predict(X_test)

xgb_accuracy = accuracy_score(y_test, xgb_predictions)

print("\nXGBoost Accuracy:", xgb_accuracy)
```

## 17. Model Comparison (Initial Models)

```python
# ==========================================================
# STEP 17 : MODEL COMPARISON
# ==========================================================

model_names = [
    'SVM',
    'KNN',
    'Random Forest',
    'XGBoost'

]

accuracies = [
    svm_accuracy,
    knn_accuracy,
    rf_accuracy,
    xgb_accuracy


]

results = pd.DataFrame({
    'Model': model_names,
    'Accuracy': accuracies
})

print("\n", results)
```

## 18. Accuracy Visualization

```python
# ==========================================================
# STEP 18 : ACCURACY VISUALIZATION
# ==========================================================

plt.figure(figsize=(8,5))
sns.barplot(x=model_names, y=accuracies)
plt.title("Model Accuracy Comparison")
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.show()
```

## 19. Best Model Selection

```python
# ==========================================================
# STEP 19 : BEST MODEL SELECTION
# ==========================================================

accuracy_dict = {
    "SVM": svm_accuracy,
    "KNN": knn_accuracy,
    "Random Forest": rf_accuracy,
    "XGBoost": xgb_accuracy
}

best_model_name = max(
    accuracy_dict,
    key=accuracy_dict.get
)

best_accuracy = accuracy_dict[best_model_name]

if best_model_name == "SVM":
    best_model = best_svm
    best_predictions = svm_predictions
elif best_model_name == "KNN":
    best_model = best_knn
    best_predictions = knn_predictions
elif best_model_name == "Random Forest":
    best_model = best_rf
    best_predictions = rf_predictions
else:
    best_model = xgb_model
    best_predictions = xgb_predictions

print("\nBest Model:", best_model_name)
print("Best Accuracy:", best_accuracy)
```

## 20. Classification Report

```python
# ==========================================================
# STEP 20 : CLASSIFICATION REPORT
# ==========================================================

print("\nClassification Report:\n")
print(classification_report(y_test, best_predictions))
```

## 21. Confusion Matrix

```python
# ==========================================================
# STEP 21 : CONFUSION MATRIX
# ==========================================================

cm = confusion_matrix(y_test, best_predictions)

plt.figure(figsize=(6,5))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues'
)

plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

## 22. Learning Curve

```python
# ==========================================================
# STEP 22 : LEARNING CURVE
# ==========================================================

train_sizes, train_scores, test_scores = learning_curve(
    best_model,
    X_train,
    y_train,
    cv=3,
    scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 5)
)

train_mean = np.mean(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)

plt.figure(figsize=(8,5))

plt.plot(
    train_sizes,
    train_mean,
    label='Training Accuracy'
)

plt.plot(
    train_sizes,
    test_mean,
    label='Validation Accuracy'
)

plt.title('Learning Curve')
plt.xlabel('Training Size')
plt.ylabel('Accuracy')
plt.legend()
plt.grid()
plt.show()
```

## 23. Sample Predictions

```python
# ==========================================================
# STEP 23 : SAMPLE PREDICTIONS
# ==========================================================

plt.figure(figsize=(12,5))

for i in range(5):
    plt.subplot(1,5,i+1)
    plt.imshow(processed_images[i])

    actual = classes[labels[i]]
    sample_feature = scaler.transform([features[i]])
    sample_feature = pca.transform(sample_feature)
    pred = best_model.predict(sample_feature)[0]
    predicted_class = classes[pred]

    plt.title(f"A:{actual}\nP:{predicted_class}")
    plt.axis('off')

plt.show()
```

```python
# ==========================================================
# FINAL CONCLUSION
# ==========================================================

print("\nPipeline Completed Successfully")
print("Best Model:", best_model_name)
print("Final Accuracy:", best_accuracy)
```
