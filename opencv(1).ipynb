```python
import cv2

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open webcam")
    exit()

print("Press 'q' to quit the webcam window")

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    cv2.imshow("Webcam Feed", frame)

    # Wait 1 ms and check if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("Quitting...")
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
camera = cv2.VideoCapture(0)
frame_width  = int (camera.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(camera.get(cv2.CAP_PROP_FRAME_HEIGHT))
codec = cv2.VideoWriter_fourcc(*'XVID')
recorder  = cv2.VideoWriter("my_video.avi",codec,20, (frame_width,frame_height))
while True:
    success,image = camera.read()
    
    if not success:
        break
    recorder.write(image)
    cv2.imshow("Rocording live",image)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
camera.release()
recorder.release()
cv2.destroyAllWindows() 
```

üß† Step 2: How Kernel Works (Convolution Example)

We‚Äôll take:
A 3√ó3 kernel (for sharpening)
A small 5√ó5 image patch
üñºÔ∏è Image Patch (Grayscale)
Let‚Äôs assume a small portion of an image with pixel intensities:
I = [
 [10, 10, 10, 10, 10],
 [10, 20, 20, 20, 10],
 [10, 20, 50, 20, 10],
 [10, 20, 20, 20, 10],
 [10, 10, 10, 10, 10]
]
üß© Kernel (3√ó3 Sharpen Filter)
K = [
 [ 0, -1,  0],
 [-1,  5, -1],
 [ 0, -1,  0]
]
‚öôÔ∏è Step-by-Step Convolution
The kernel slides over the image, centered at each pixel (excluding borders).
Let‚Äôs take the center pixel (value = 50) and apply the kernel to its 3√ó3 neighborhood:
Neighborhood = [
 [20, 20, 20],
 [20, 50, 20],
 [20, 20, 20]
]
üßÆ Multiply Element-wise
We multiply each kernel element by its corresponding pixel value:
[ 0*20,  -1*20,  0*20 ],
[-1*20,   5*50, -1*20 ],
[ 0*20,  -1*20,  0*20 ]
Result after multiplication:
[   0,  -20,    0 ],
[ -20,  250,  -20 ],
[   0,  -20,    0 ]
‚ûï Sum All Values
Now, add all elements together:
0 - 20 + 0 - 20 + 250 - 20 + 0 - 20 + 0 = 170
üßæ Replace Center Pixel
So, the new pixel value = 170 (originally it was 50).
üëâ This means the bright spot became even brighter ‚Üí the image is sharpened.
üìä Intuition
Kernel gives positive weight (5) to the center pixel ‚Üí emphasizes it.
Surrounding pixels get negative weights (-1) ‚Üí reduce their influence.
This contrast boosts edges and details.
üí° Note
Blurring kernels do the opposite (average the neighborhood).

Edge detection kernels emphasize differences between neighbors.

```python
#sharpeining
import cv2
import numpy as np 
image2 = cv2.imread("Screenshot 2025-10-22 091456.png")
sharpen_kernel = np.array([
    [0,-1,0],
    [-1,5,-1],
    [0,-1,0]
    
])
sharpened = cv2.filter2D(image2,-1,sharpen_kernel)
cv2.imshow("Original Image",image2)
cv2.imshow("Sharpened Image",sharpened)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
